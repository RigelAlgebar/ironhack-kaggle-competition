```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
solar_train = pd.read_csv('../data/external/solar_train.csv')
```

```python
solar_train.head()
```

```python
solar_train.describe()
```

```python
solar = solar_train.copy(deep=True)
```

```python
def temp_to_kelvin(t):
    return round((t-32)*5/9 + 273.15, 2)
```

```python
def press_to_pascal(p):
    return round(p*133.322, 2)
```

```python
def speed_to_ms(s):
    return round(s/2.236936, 2)
```

```python
solar['Temperature'] = solar['Temperature'].apply(temp_to_kelvin)
solar['Pressure'] = solar['Pressure'].apply(press_to_pascal)
solar['Speed'] = solar['Speed'].apply(speed_to_ms)
```

```python
solar.describe()
```

```python
solar.head()
```

```python
solar.to_csv('../data/interim/solar_int_train.csv')
```

```python
solar_phys = solar.copy(deep=True)
```

```python
solar_phys.drop(columns=['UNIXTime', 'Data', 'Time', 'TimeSunRise', 'TimeSunSet'], inplace=True)
```
